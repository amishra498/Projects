# Read Dataset

```python
import pandas as pd
A = pd.read_csv("C:/Users/JOHN/Downloads/training_set (1).csv")
A.head() # for preview
```

```python
A.info()
```

```python
A.shape
```

# Missing data treatment

```python
A.isna().sum()
```

```python
for i in A.columns:
    if(A[i].dtypes == "object"):
        x = A[i].mode()[0]
        A[i] = A[i].fillna(x)
    else:
        x = A[i].mean()
        A[i] = A[i].fillna(x)
```

```python
A.isna().sum()
```

# Remove Unneccessary columns

```python
A = A.drop(labels=["Loan_ID"],axis=1) # Loan_ID contains lot of unique values.
```

```python
A.head()
```

# defie X and Y

```python
Y = A[["Loan_Status"]]
X = A.drop(labels=["Loan_Status"],axis=1)
```

```python
Y.head(5)
```

```python
X.head()
```

# Preprocessing of X and Y columns:

```python
cat = []
con = []
for i in X.columns:
    if(X[i].dtypes == 'object'):
        cat.append(i)
    else:
        con.append(i)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler() # continous data converted into same magnitude.
x1 = pd.DataFrame(ss.fit_transform(X[con]),columns=con)
x2 = pd.get_dummies(X[cat]) # categorical columns converted into 0 and 1

Xnew = x1.join(x2)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Ynew = pd.DataFrame(le.fit_transform(Y),columns = ["Loan_Status"])
```

```python
Xnew
```

# divide data into training and testing set

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(Xnew,Ynew,test_size=0.2,random_state=21)
```

# Create a Decision Tree Classification Model

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
model = dtc.fit(xtrain,ytrain)
tr_pred = model.predict(xtrain)
ts_pred = model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
tr_acc = accuracy_score(tr_pred,ytrain)
ts_acc = accuracy_score(ts_pred,ytest)
```
