```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
scaled_tr_data = mm.fit_transform(tr_data)
scaled_tr_data
```

# Create data for training  in such way that 50 rows as input and 51 row will be output

```python
tr_data.head()
```

```python
tr_data.shape
```

```python
xtrain = []
ytrain = []
for i in range(50,1258,1):
    xtrain.append(tr_data[i-50:i]) # 50 rows as input
    ytrain.append(tr_data[i:i+1]) # 51 row will be output
```

```python
print(xtrain[0],"\n") 
print(ytrain[0])
```

```python
xtr = np.array(xtrain)# Neural Network always treated data in the form of array
ytr = np.array(ytrain)
```

# Create a Recurrent Neural Network Model

```python
from keras.models import Sequential
from keras.layers import LSTM,Dense,Dropout
from keras.callbacks import EarlyStopping
```

```python
es = EarlyStopping(patience=5,monitor = "loss")
nn = Sequential()
nn.add(LSTM(50,return_sequences = True,input_shape = (50,1)))
nn.add(Dense(100))
nn.add(Dense(1))
nn.compile(optimizer="adam",loss = "mean_squared_error")
nn.fit(xtr,ytr,epochs=10000,callbacks=es)
```
