# Unify cases and remove special character

```python
Q = []
from re import sub
for i in A.Review:
    Q.append(sub("[^A-Za-z,0-9 ]","",i.upper()))

#special character removed and all the charecter replaced in upper case
```

```python
Q
```

# Bag of words or Count vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
word_vect = cv.fit_transform(Q).toarray()
```

```python
words = cv.get_feature_names() # all unique words
```

```python
# define X and Y
X = word_vect
Y = A[["Liked"]]

# divide data into trainning and testing set
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=21)
```

```python
X.shape
```

```python
xtrain.shape
```

```python
xtest.shape
```

# Artificial Neural Network model

```python
from keras.models import Sequential # adding one layer at a time
from keras.layers import Dense,Dropout,Flatten

nn = Sequential()

nn.add(Dense(100,activation = 'relu',input_dim = (2067)))
nn.add(Dropout(0.8))

nn.add(Dense(100))
nn.add(Dropout(0.8))

nn.add(Dense(1,activation = "sigmoid"))
```

```python
# compilation
nn.compile(loss = "binary_crossentropy",metrics = ["accuracy"])
nn.fit(xtrain,ytrain,validation_split=0.2,epochs=30)
```

# Testing and Evaluation

```python
W = []
for i in nn.predict(xtest):
    if(i[0] > 0.5):
        W.append(1)
    else:
        W.append(0) 
```

```python
W
```

```python
from sklearn.metrics import accuracy_score
test_accuracy = accuracy_score(W,ytest)
```
