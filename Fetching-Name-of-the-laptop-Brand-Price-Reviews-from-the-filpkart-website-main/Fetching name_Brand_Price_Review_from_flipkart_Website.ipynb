# importing important libraries

```python
from bs4 import BeautifulSoup
from requests import get
```

```python
data = get(url)
data
```

```python
data.content # HTML is not in proper formate
```

```python
soup = BeautifulSoup(data.content)# we get proper html
```

```python
soup
```

# Firstly fetch name of Laptop

```python
soup.find_all("h1")
```

```python
name = []
for i in soup.find_all("h1"):
    name.append(i.span.text)
name
```

# fetching brand

```python
brand = []
brand.append(name[0][0:6])
```

```python
brand
```

# Fetching Price

```python
price = []
for i in soup.find_all("div",class_ = ["_30jeq3 _16Jk6d"]):
    price.append(i.text)
    
price
    
```

# Fetchinng Reviews

```python
reviews = []
Q = []
for i in soup.find_all("span",class_ = ["_2_R_DZ"]):
    Q.append(i.text)
```

```python
reviews.append(Q[0][16:19])
```
